<h3>First-Order Logic (FOL)</h3>

```python
rules = {
    "FeverAndCough_implies_Flu": ([("Fever", "x"), ("Cough", "x")], ("Flu", "x")),
    "Flu_implies_NeedsRest": ([("Flu", "x")], ("NeedsRest", "x")),
    "SevereCoughAndBreath_implies_Pneumonia": ([("SevereCough", "x"), ("ShortnessOfBreath", "x")], ("Pneumonia", "x"))
}

facts = {("Fever", "John"), ("Cough", "John")}
proofs = {f: "ASSERTED" for f in facts}

def substitute(term, const):
    return tuple(const if t == "x" else t for t in term)

def forward_chain():
    iteration = 1
    changed = True
    while changed:
        print(f"[FC] Iteration {iteration}, facts={len(facts)}")
        changed = False
        for rule, (premises, conclusion) in rules.items():
            for const in {c for _, c in facts}:
                prem = [substitute(p, const) for p in premises]
                concl = substitute(conclusion, const)
                if all(p in facts for p in prem) and concl not in facts:
                    facts.add(concl)
                    proofs[concl] = (rule, prem)
                    print(f"[INFER] {concl} via {rule} using {prem}")
                    changed = True
        iteration += 1

def explain(query):
    print(f"\n=== QUERY: {query[0]}({query[1]}) ===")
    if query not in facts:
        print("Known: False")
        print(f"{query[0]}({query[1]}) is NOT known.")
        return
    if proofs[query] == "ASSERTED":
        print("Known: True")
        print(f"{query[0]}({query[1]}) is an asserted fact.")
    else:
        rule, premises = proofs[query]
        print("Known: True")
        print(f"{query[0]}({query[1]}) was INFERRED by rule '{rule}'.")
        print("Support facts:")
        for p in premises:
            explain(p)

forward_chain()
explain(("Flu", "John"))
explain(("NeedsRest", "John"))
explain(("Pneumonia", "John"))
```
