<h4>Handwritten Digit Recognition (MNIST Dataset)-Implement a 3-layer neural network
(Input → Hidden → Output) using Backpropagation.
    
1. Use Sigmoid, ReLU, or Tanh activation functions.
2. Train the network on the MNIST dataset to classify handwritten digits (0-9).
3. Optimize using Gradient Descent with Momentum or Adam Optimizer.</h4>

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
x_test.shape
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train.shape
x_test.shape
```

```python
index=0
plt.imshow(x_train[index])
print(y_train[index])
```

<h3>Relu Activation Function</h3>

```python
model1 = Sequential([
    Flatten(input_shape=(28, 28)),     
    Dense(128, activation='relu'),    
    Dense(10, activation='softmax')])
```

```python
model1.compile(optimizer=Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model1.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model1.evaluate(x_test, y_test)
print(f"\nTest Accuracy (Relu): {test_acc:.4f}")
```

```python
plt.figure(figsize=(12,4))
    
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.title('Loss - Relu')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
    
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.title('Loss - Relu')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
    
plt.show()
```

<h3>Sigmoid Activation Function</h3>

```python
model2 = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='sigmoid'),
    Dense(10, activation='softmax')])
```

```python
model2.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history2 = model2.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model2.evaluate(x_test, y_test)
print(f"Test Accuracy (Sigmoid): {test_acc:.4f}")
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history2.history['accuracy'], label='Train Accuracy')
plt.plot(history2.history['val_accuracy'], label='Test Accuracy')
plt.title('Accuracy - Sigmoid')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history2.history['loss'], label='Train Loss')
plt.plot(history2.history['val_loss'], label='Test Loss')
plt.title('Loss - Sigmoid')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

<h3>Tanh Activation Function</h3>

```python
model3 = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='tanh'),
    Dense(10, activation='softmax')])
```

```python
history3 = model3.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```
