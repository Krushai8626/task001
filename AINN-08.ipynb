<h4>Clustering 2D Data using Winner-Takes-All (WTA) Network

1. Implement a Winner-Takes-All (WTA) neural network to cluster 2D data points.
2. Each neuron represents a cluster center, and only the winning neuron (closest to input)
updates its weights.
3. Train the network on randomly generated 2D points and visualize the clusters.
</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h5> Step 1: Generate random 2D input data
</h5>

```python
np.random.seed(42)
n_points = 200
data = np.random.rand(n_points, 2) * 10
```

<h5> Step 2: Initialize weights (cluster centers)
</h5>

```python
n_clusters = 4
weights = np.random.rand(n_clusters, 2) * 10
learning_rate = 0.2
epochs = 20
```

<h5> Step 3: Train the WTA network
</h5>

```python
for epoch in range(epochs):
    for x in data:
        distances = np.sqrt(((x[0] - weights[:, 0]) ** 2) + ((x[1] - weights[:, 1]) ** 2))
        winner = np.argmin(distances)
        weights[winner][0] += learning_rate * (x[0] - weights[winner][0])
        weights[winner][1] += learning_rate * (x[1] - weights[winner][1])
    
    learning_rate *= 0.95
```

<h5> Step 4: Assign each point to the nearest cluster
</h5>

```python
labels = []
for x in data:
    distances = np.sqrt(((x[0] - weights[:, 0]) ** 2) + ((x[1] - weights[:, 1]) ** 2))
    labels.append(np.argmin(distances))
labels = np.array(labels)
```

<h5> Step 5: Visualization of clusters
</h5>

```python
plt.figure(figsize=(8,5))
for i in range(n_clusters):
    plt.scatter(data[labels == i, 0], data[labels == i, 1], s=25, label=f'Cluster {i+1}')
plt.scatter(weights[:, 0], weights[:, 1], c='black', marker='*', s=150, label='Final Cluster Centers')
plt.title("2D Clustering using Winner-Takes-All (WTA) Neural Network")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.grid(True)
plt.show()
```
